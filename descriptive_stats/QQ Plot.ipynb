```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
n = 1000
data = np.random.randn(n)

x = np.linspace(-4, 4, 10001)

theonorm = stats.norm.pdf(x)
theonorm = theonorm / max(theonorm)

yy, xx = np.histogram(data, 40)
yy = yy / max(yy)
xx1 = (xx[: -1] + xx[1 : ]) / 2
```

```python
plt.figure(figsize = (15, 8))

plt.plot(xx1, yy, label = 'Empirical')
plt.plot(x, theonorm, label = 'Theoretical')
plt.legend()
plt.show()
```

```python
zSortData = np.sort(stats.zscore(data))
sortNorm = stats.norm.ppf(np.linspace(0, 1, n))

plt.figure(figsize = (12, 8))
plt.plot(zSortData, '.', label = 'z score')
plt.plot(sortNorm, '.', label = 'sort normal')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 10))
plt.plot(sortNorm, zSortData, '.')

xL, xR = plt.xlim()
yL, yR = plt.ylim()
lims = [np.min([xL, xR, yL, yR]), np.max([xL, xR, yL, yR])]
print([xL, xR, yL, yR])
plt.xlim(lims)
plt.ylim(lims)

plt.plot(lims, lims)

plt.show()
```
