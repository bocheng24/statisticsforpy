```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
N = 10001

d1 = np.random.randn(N) - 1
d2 = np.random.randn(N) * 3
d3 = np.random.randn(N) + 1
```

```python
d1
```

```python
d2
```

```python
d3
```

```python
nbins = 30

y1, x1 = np.histogram(d1, nbins)
xx1 = (x1[1 :] + x1[: -1]) / 2

y2, x2 = np.histogram(d2, nbins)
xx2 = (x2[1 :] + x2[: -1]) / 2

y3, x3 = np.histogram(d3, nbins)
xx3 = (x3[1 :] + x3[: -1]) / 2

plt.figure(figsize = (15, 8))

plt.plot(xx1, y1, '#ED4C67', label = 'd1')
plt.plot(xx2, y2, '#1289A7', label = 'd2')
plt.plot(xx3, y3, '#D980FA', label = 'd3')

plt.xlabel('Value counts')
plt.ylabel('Data values')

plt.legend()
plt.show()
```

```python
mean_d1 = sum(d1) / len(d1)
mean_d2 = np.mean(d2)
mean_d3 = np.mean(d3)

plt.figure(figsize = (15, 8))

plt.plot(xx1, y1, '#ED4C67', label = 'd1')
plt.plot(xx2, y2, '#1289A7', label = 'd2')
plt.plot(xx3, y3, '#D980FA', label = 'd3')

plt.plot([mean_d1, mean_d1], [0, max(y1)], 'r--')
plt.plot([mean_d2, mean_d2], [0, max(y2)], 'b--')
plt.plot([mean_d3, mean_d3], [0, max(y3)], 'g--')

plt.xlabel('Value counts')
plt.ylabel('Data values')

plt.legend()
plt.show()
```

```python
d4 = np.hstack((np.random.randn(N) - 2, np.random.randn(N) + 2))
```

```python
y4, x4 = np.histogram(d4, nbins)
xx4 = (x4[1 :] + x4[: -1]) / 2
mean_d4 = np.mean(d4)

plt.figure(figsize = (15, 8))
plt.plot(xx4, y4)
plt.plot([mean_d4, mean_d4], [0, max(y4)])

plt.xlabel('Value counts')
plt.ylabel('Data values')

plt.show()
```

```python
shift = 0
stretch = 0.7
n = 2000
nbins = 50

data = stretch * np.random.randn(n) + shift
data = np.exp(data)

y, x = np.histogram(data, nbins)
xx = (x[1 :] + x[: -1]) / 2

data_mean = np.mean(data)
data_median = np.median(data)
```

```python
fig, ax = plt.subplots(2, 1, figsize = (15, 10))

ax[0].plot(data, '.', color = '#f6b93b', alpha = 0.45, label = 'data')
ax[0].plot([1, n], [data_mean, data_mean], 'r--', label = 'Mean')
ax[0].plot([1, n], [data_median, data_median], 'b--', label = 'Median')
ax[0].legend()

ax[1].plot(xx, y, color = '#f6b93b', label = 'data')
ax[1].plot([data_mean, data_mean], [0, max(y)], 'r--', label = 'Mean')
ax[1].plot([data_median, data_median], [0, max(y)], 'b--', label = 'Median')
ax[1].legend()

plt.show()
```
