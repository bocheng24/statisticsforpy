```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
N = 10001

x = np.linspace(-4, 4, N)
gausdist = stats.norm.pdf(x)

plt.figure(figsize = (12, 7))
plt.plot(x, gausdist)
plt.title('Gaussian | Normal distribution')
plt.show()
```

```python
sum(gausdist)
```

```python
plt.figure(figsize = (12, 7))
plt.plot(x, gausdist / sum(gausdist))
plt.title('Gaussian | Normal distribution')
plt.show()

print(sum(gausdist / sum(gausdist)))
```

```python
width = np.diff(x[: 2])
areaForeach = gausdist * width
bell_area = sum(areaForeach)
```

```python
# Normally distributed for random number

variance = 1 
mean = 5 
n = 1000

data = variance * np.random.randn(n) + mean

plt.figure(figsize = (12, 7))
plt.hist(data, bins = 25)
plt.title('Empirical normal distribution')
plt.show()

print(np.mean(data))
```

```python
# Uniformly distributed numbers

variance = 2
mean = 0.5
n = 10000

data = variance * np.random.rand(n) + mean - variance / 2

fig, ax = plt.subplots(2, 1, figsize = (15, 12))

ax[0].plot(data, '.', markersize = 1.2)
ax[0].set_title('Uniform data values')

ax[1].hist(data, 25)
ax[1].set_title('Uniform data histgram')

plt.show()
```

```python
# log normal distribution

N = 1001

x = np.linspace(0, 10, N)
lognormal = stats.lognorm.pdf(x, 1)

plt.figure(figsize = (12, 7))
plt.plot(x, lognormal)
plt.title('Log Normal distribution')
plt.show()
```

```python
# Empirical log normal distribution

variance = 5
mean = 0.5
n = 2000

data = mean * np.random.randn(n) + variance
data = np.exp(data)

fig, ax = plt.subplots(2, 1, figsize = (15, 12))

ax[0].plot(data, '.', markersize = 5.5)
ax[0].set_title('Log normalize data values')

ax[1].hist(data, 25)
ax[1].set_title('Log normalize data histgram')

plt.show()
```

```python
# Binomial

n = 10
p = 0.5

x = range(n + 2)
binodist = stats.binom.pmf(x, n, p)

plt.figure(figsize = (12, 7))
plt.bar(x, binodist)
plt.title(f'Binomial distribution (n = {n}, p = {p})')
plt.show()
```

```python
# t distribution

x = np.linspace(-5, 5, 1001)
df = 2 # degrees of freedom

t = stats.t.pdf(x, df)

plt.figure(figsize = (12, 7))
plt.plot(x, t)
plt.xlabel('t value')
plt.ylabel('P (t | H$_0$)')
plt.title(f'T ( %g ) distribution'%df)
plt.show()
```

```python
# F distribution

num_df = 5
den_df = 100

x = np.linspace(0, 10, 10001)
fdist = stats.f.pdf(x, num_df, den_df)

plt.figure(figsize = (12, 7))
plt.plot(x, fdist)
plt.xlabel('f value')
plt.title(f'F ({num_df}, {den_df}) distribution')
plt.show()
```
