```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
N = 1000
numbers = np.ceil(8 * np.random.randn(N) ** 2)

u = np.unique(numbers)
probs = np.zeros(len(u))
```

```python
for ui in range(len(u)):
    probs[ui] = sum(numbers == u[ui]) / N
```

```python
probs
```

```python
entropy = -sum(probs * np.log2(probs + np.finfo(float).eps))
```

```python
entropy
```

```python
plt.figure(figsize = (12, 8))
plt.bar(u, probs)
plt.title('Entropy = %g'%entropy)
plt.xlabel('Unique number')
plt.ylabel('Probablilties')
plt.show()
```

```python
N = 1123
brownNoise = np.cumsum(np.sign(np.random.randn(N)))

fig, ax = plt.subplots(2, 1, figsize = (12, 10))
ax[0].plot(brownNoise)
ax[0].set_title('Brownian noise')
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data Value')

ax[1].hist(brownNoise / sum(brownNoise), 30)
ax[1].set_xlabel('Data Value')
ax[1].set_ylabel('Probabilites')
plt.show()
```

```python
nbins = [10, 20, 30, 40, 50]

for b in nbins:
    nPerBin, bins = np.histogram(brownNoise, b)
    probs = nPerBin / sum(nPerBin)

    entropy = -sum(probs * np.log2(probs + np.finfo(float).eps))
    print(entropy)
```
