```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rows = 30
cols = 6

data = np.zeros((rows, cols))
data
```

```python
for i in range(cols):
    
    data[:, i] = 30 * np.random.randn(rows) * (2 * i / (cols - 1)) ** 2 + (i + 1) ** 2

# data
```

```python
plt.figure(figsize = (12, 8))

plt.boxplot(data)
plt.show()
```

```python
plt.figure(figsize = (12, 8))

plt.boxplot(data.T)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(data = data, orient = 'v')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(data = data.T)
plt.show()
```

```python
# Gaussian distribution
data1 = np.random.randn(rows)

# Uniform distribution
data2 = np.random.rand(rows)

df = pd.DataFrame(np.array([data1, data2]).T, columns = ['Gaussian distribution', 'Uniform distribution'])
df.head()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(data = df)

plt.show()
```
