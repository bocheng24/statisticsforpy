```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1 / 2), np.log10(nbins - 0.01), totalN))
```

```python
rawdata
```

```python
uniqueNums = np.unique(rawdata)
dataPie = np.zeros(len(uniqueNums))
```

```python
for i in range(len(uniqueNums)):
    dataPie[i] = sum(rawdata == uniqueNums[i])
```

```python
dataPie
```

```python
plt.figure(figsize = (12, 8))
plt.pie(x = dataPie, labels = 100 * dataPie / sum(dataPie))
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.pie(x = dataPie, 
        labels = ['zero', 'one', 'two', 'three', 'four'],
        explode = [0.02, 0.02, 0.02, 0.2, 0.02]
       )
plt.show()
```
