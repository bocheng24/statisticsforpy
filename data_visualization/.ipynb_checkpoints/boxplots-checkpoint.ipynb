```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rows = 30
cols = 6

data = np.zeros((rows, cols))
data
```

```python
for i in range(cols):
    
    data[:, i] = 30 * np.random.randn(rows) * (2 * i / (cols - 1)) ** 2 + (i + 1) ** 2

# data
```

```python
fig, ax = plt.subplots(1, 3, figsize = (16, 4))

ax[0].bar(range(cols), np.mean(data, axis = 0))
ax[0].set_title('bar plot')

ax[1].errorbar(range(cols), np.mean(data, axis = 0), np.std(data, axis = 0, ddof = 1), marker = 's', linestyle = '')
ax[1].set_title('error bar plot')

ax[2].bar(range(cols), np.mean(data, axis = 0))
ax[2].errorbar(range(cols), np.mean(data, axis = 0), np.std(data, axis = 0, ddof = 1), marker = 's', linestyle = '')
ax[2].set_title('bar + error bar plot')

plt.show()
```

```python
xcrossings = [1, 2, 4, 5, 6, 9]

plt.figure(figsize = (12, 8))

plt.bar(xcrossings, np.mean(data, axis = 0), color = '#341f97')
plt.errorbar(xcrossings, 
             np.mean(data, axis = 0), 
             np.std(data, axis = 0, ddof = 1), 
             marker = 's', 
             linestyle = '',
             color = '#ee5253')

plt.xticks(ticks = xcrossings)

plt.title('bar + error bar plot')

plt.show()
```

```python
data2 = [
    [2, 5, 4, 3],
    [1, 2, 8, 5]
]

fig, ax = plt.subplots(2, 2, figsize = (16, 12))

ax[0, 0].imshow(data2)
ax[0, 1].imshow(np.array(data2).T)

df = pd.DataFrame(data2, columns = ['p0', 'p1', 'p2', 'p3'])

df.plot(kind = 'bar', ax = ax[1, 0])
df.T.plot(kind = 'bar', ax = ax[1, 1])

plt.show()
```
