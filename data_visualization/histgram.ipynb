```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000

# log normal distribution
data = np.exp(np.random.randn(n) / 2)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(data, marker = '.', linestyle = '')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.hist(x = data, bins = 40)
plt.show()
```

```python
y, x = np.histogram(data, bins = 40)

xx = (x[1:] + x[:-1]) /2

plt.figure(figsize = (12, 8))
plt.plot(xx, y)
plt.show()
```

```python
y_pct = [round(val / sum(y) * 100, 2) for val in y]
xticks = [round(v, 2) for v in x]

plt.figure(figsize = (12, 8))

plt.hist(x = data, bins = 40, density = True)
plt.xticks(ticks = xticks, rotation = 45)

plt.show()
```
