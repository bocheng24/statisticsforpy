```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n = 1000

# arbitrary hist bins
k = 40

data = np.exp(np.random.randn(n) / 2)

plt.figure(figsize = (12, 7))
plt.hist(data, k)
plt.xlabel('value')
plt.ylabel('counts')
plt.show()
```

```python
# Freedman-diaconis rule bin

r = 2 * stats.iqr(data) * n ** (-1 / 3)
b = np.ceil((max(data) - min(data)) / r)

print(f'r: {r}')
print(f'b: {b}')
```

```python
plt.figure(figsize = (12, 7))
plt.hist(data, int(b))
plt.xlabel('value')
plt.ylabel('counts')
plt.show()
```

### Built-in F-D bins

```python
plt.figure(figsize = (12, 7))
plt.hist(data, bins = 'fd')
plt.xlabel('value')
plt.ylabel('counts')
plt.show()
```
