```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 40

dx = np.exp(-abs(np.random.randn(N) * 3))
dy = np.exp(-abs(np.random.randn(N) * 5))

# mean for x and mean for y
datamean = [np.mean(dx), np.mean(dy)]

ds = np.zeros(N)

# Square distance for both x and y
for i in range(N):
    
    ds[i] = np.sqrt((dx[i] - datamean[0]) ** 2 + (dy[i] - datamean[1]) ** 2)
    
ds = (ds - np.mean(ds)) / np.std(ds)
```

```python
ds
```

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 8))

ax[0].plot(dx, dy, 'ko', markerfacecolor = 'k')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_xlabel('Variable X')
ax[0].set_ylabel('Variable Y')

ax[0].plot(datamean[0], datamean[1], 'kp', markerfacecolor = 'g', markersize = 15)

ax[1].plot(ds, 'ko', markerfacecolor = [0.7, 0.5, 0.3], markersize = 12)
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Z distance')

plt.show()
```

```python
distanceThresh = 2.5

print(np.where(ds > distanceThresh))
oidx = np.where(ds > distanceThresh)[0]
print(oidx)

ax[1].plot(oidx, ds[oidx], 'x', color = 'r', markersize = 20)
ax[0].plot(dx[oidx], dy[oidx], 'x', color = 'r', markersize = 20)

fig
```
