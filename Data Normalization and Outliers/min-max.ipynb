```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 42
data = np.log(np.random.rand(N)) * 234 + 934

datamin = min(data)
datamax = max(data)

data_range = datamax - datamin
data_scale = (data - datamin) / data_range
```

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 8))
ax[0].plot(1 + np.random.randn(N) / 20, data, 'ks')
# ax[0].set_xlim([0, 2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original Data')


ax[1].plot(1 + np.random.randn(N) / 20, data_scale, 'ks')
# ax[1].set_xlim([0, 2])
ax[1].set_xticks([])
ax[1].set_ylabel('Min-max scale')
ax[1].set_title('Min-max Data')
plt.show()
```

```python
plt.figure(figsize = (12, 7))
plt.plot(data, data_scale, 'ks')
plt.xlabel('Original')
plt.ylabel('Min-max')
plt.show()
```

```python
newMin = 2
newMax = 6

data_shift = data_scale * (newMax - newMin) + newMin
[min(data_shift), max(data_shift)]
```

```python
plt.figure(figsize = (12, 7))
plt.plot(data, data_scale, '.', markersize = 7.5, label = 'min-max')
plt.plot(data, data_shift, '.', markersize = 7.5, label = 'shifted min-max')
plt.xlabel('Original')
plt.ylabel('Min-max')
plt.legend()
plt.show()
```
