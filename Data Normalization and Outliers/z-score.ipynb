```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = np.random.poisson(3, 1000) ** 2

datamean = np.mean(data)
datastd = np.std(data, ddof = 1)

plt.figure(figsize = (12, 7))

plt.plot(data, '.', markersize = 7.5)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Mean: {np.round(datamean, 2)} | STD: {np.round(datastd, 2)}')
plt.show()
```

```python
dataz = (data - datamean) / datastd
dataz[: 10]
```

```python
dataz = stats.zscore(data)
dataz[: 10]
```

```python
dataZmean = np.mean(dataz)
dataZstd = np.std(dataz, ddof = 1)

plt.figure(figsize = (12, 7))
plt.plot(dataz, '.', markersize = 7.5)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Z Mean: {np.round(dataZmean, 2)} | Z STD: {np.round(dataZstd, 2)}')
plt.show()
```

```python
plt.figure(figsize = (12, 7))
plt.plot(data, dataz, '.')
plt.xlabel('Original')
plt.ylabel('Z-Transformed')
plt.title(f'Correlation r = {np.corrcoef(data, dataz)[0, 0]}')
plt.show()
```
